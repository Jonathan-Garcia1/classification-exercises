- Write a function to split your data into train, test and validate datasets. Add this function to prepare.py.


```python
#import at start
from sklearn.model_selection import train_test_split
```

```python
def splitf (df, param_stratify, param_seed):
    seed = param_seed

    train, val_test = train_test_split(df, train_size = 0.7,
                                    random_state = seed,
                                    stratify = df[param_stratify])

    val, test = train_test_split(val_test, train_size = 0.5,
                                random_state = seed,
                                stratify = val_test[param_stratify])
    return train, val, test
```

<div style='background-color: #F5F5F5; padding-left: 40px; padding-top: 10px; padding-bottom: 10px; padding-right: 10px; color: #1A5E7F; font-size: 18px; box-sizing: border-box;'>

- For virginica & versicolor: Compare the mean petal_width using the Mann-Whitney test (scipy.stats.mannwhitneyu) to see if there is a significant difference between the two groups. Do the same for the other measurement types.

<div style='background-color: #F5F5F5; padding-left: 40px; padding-top: 10px; padding-bottom: 10px; padding-right: 10px; color: Green; font-size: 18px; box-sizing: border-box;'>
Null: There is no significant difference between Versicolor and Virginica petal width
Alternative: There is a significant difference between Versicolor and Virginica petal width

```python
from scipy import stats

# Select data for "versicolor" and "virginica" species
versicolor_data = iris_train[iris_train['species'] == 'versicolor']
virginica_data = iris_train[iris_train['species'] == 'virginica']

s_value, p_value = stats.mannwhitneyu(versicolor_data['petal_width'], virginica_data['petal_width'], alternative='two-sided')

print(f'Test for petal_width')
print(f's_value: {s_value}')
print(f'p-value: {p_value}')

if p_value < 0.05:
    print("Reject the Null (p < 0.05)\n")
else:
    print("Fail to reject Null (p >= 0.05)\n")
```

Sigificant difference in all measurements between Versicolor and Virginica species
